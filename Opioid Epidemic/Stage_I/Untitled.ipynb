#### Calculate mean and median death rate at state level.


```python
State_Mean = Merged_df.groupby('State')['Norm_Deaths'].mean().reset_index()
State_Mean.columns = ['State','Norm_Deaths_Mean']
State_Median = Merged_df.groupby('State')['Norm_Deaths'].median().reset_index()
State_Median.columns = ['State','Norm_Deaths_Median']
State_Info = pd.merge(State_Median,State_Mean, on = 'State', how = 'inner')
State_Info.head()
```

#### Merge State_Info data amd Merged_df to compare county death rate with overall state death rate.


```python
Merged_df = pd.merge(State_Info,Merged_df,on = 'State', how = 'inner')
Merged_df.head()
```

#### Find number of counties in each sate.


```python
County_Count = Merged_df.groupby('State').count()['FIPS'].reset_index()
County_Count.columns = ['State','County Count']
County_Count.head()
```

#### Find counties with death rate above state mean death rate.


```python
Merged_df_above_mean = Merged_df[Merged_df.Norm_Deaths>= Merged_df.Norm_Deaths_Mean ]
County_Count_Above_Mean = Merged_df_above_mean.groupby('State').count()['FIPS'].reset_index()
County_Count_Above_Mean.columns = ['State','Counties Above Mean State Death Rate']
County_Count_Above_Mean.head()
```

#### Find counties with death rate above state median death rate.


```python
Merged_df_above_median = Merged_df[Merged_df.Norm_Deaths>= Merged_df.Norm_Deaths_Median ]
County_Count_Above_Median = Merged_df_above_median.groupby('State').count()['FIPS'].reset_index()
County_Count_Above_Median.columns = ['State','Counties Above Median State Death Rate']
County_Count_Above_Median.head()
```

#### merge three above data frames:
#### I did not know how to merge four data frames at the same time.
#### https://stackoverflow.com/questions/23668427/pandas-three-way-joining-multiple-dataframes-on-columns


```python
import functools as ft
dfs = [County_Count,County_Count_Above_Mean,County_Count_Above_Median,State_Info]
State_Info_final = ft.reduce(lambda left, right: pd.merge(left, right, on='State'), dfs)
State_Info_final.head()
```

#### States with most counties above state mean death rate:


```python
State_Info_final[ State_Info_final['Counties Above Mean State Death Rate'] / State_Info_final['County Count'] > 0.5]
```

#### Only in few sates most counties death rate is greater than state mean death rate.
#### States with most counties above state median death rate:

```python
State_Info_final[ State_Info_final['Counties Above Median State Death Rate'] / State_Info_final['County Count'] > 0.5]
```
